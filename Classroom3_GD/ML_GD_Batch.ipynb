```python
#IMPORTING STUFFS
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#CREATING RANDOM DATASETS
x = np.random.uniform(1,101,100)     
xx = x.reshape(-1,1)
c = np.random.uniform(1,50,100)
y = np.array([0 for i in range(100)])
m = 2
for i in range(len(x)):
   y[i] = m*xx[i] + c[i]


plt.scatter(xx,y)
plt.show()
```

```python
#Just to check my GD Algo is working fine with linear regreesion. If (working fine): same results
model = LinearRegression()
model.fit(xx,y)
pred = model.predict(xx)

plt.scatter(xx,y)
plt.plot(x,pred)
plt.show()

print(f"coef : {model.coef_}, Intercept : {model.intercept_}")
```

```python
#CODE FOR BATCH GRADIENT DESCEND

def GDBatch(x,y):
    N = len(x)
    mcurr = 0
    bcurr = 0
    epoch = 100000
    lr = 0.0001
    for i in range(epoch):
        ypred = mcurr*x + bcurr
        derivm = -(2/N)*sum(x*(y-ypred))
        derivb = -(2/N)*sum(y-ypred)
        mcurr = mcurr - lr*derivm
        bcurr = bcurr - lr*derivb
    return (mcurr,bcurr)
  
mcurrr,bcurrr = GDBatch(x,y)
print(f"m : {mcurrr}, b : {bcurrr}")        
#Results matches with Linear Regression model. GD Algo seems fine :')

plt.scatter(x,y)
plt.plot(x,mcurrr*x + bcurrr,'g')
plt.show()

 
```
