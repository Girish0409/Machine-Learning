```python
#IMPORTING STUFFS
from numpy import random
from matplotlib import pyplot as plt


```

```python
#CREATING RANDOM DATASET
x = random.uniform(1,100,100)
c = random.uniform(1,70,100)
y = []
m = 10

```

```python
for i in range(0,100):
    ytemp = m*x[i] + c[i]
    y.append(ytemp)

```

```python
#PLOTTING
plt.scatter(x,y)
plt.title("Some random Scatter plot")
plt.xlabel("X Name")
plt.ylabel("Y Name")
plt.show()
```
